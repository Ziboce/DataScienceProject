```python
def log_progress(sequence, every=None, size=None, name='Items'):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{name}: {index} / ?'.format(
                        name=name,
                        index=index
                    )
                else:
                    progress.value = index
                    label.value = u'{name}: {index} / {size}'.format(
                        name=name,
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = "{name}: {index}".format(
            name=name,
            index=str(index or '?')
        )
```

```python
from PIL import Image

def crop_center(image, target_width=600, target_height=600):
    """
    Crop the image to the specified target width and height, keeping the same center.

    Args:
        image (PIL.Image): The input image.
        target_width (int): Target width of the cropped image. Default is 600.
        target_height (int): Target height of the cropped image. Default is 600.

    Returns:
        PIL.Image: The cropped image.
    """
    width, height = image.size
    left = (width - target_width) / 2
    top = (height - target_height) / 2
    right = (width + target_width) / 2
    bottom = (height + target_height) / 2
    return image.crop((left, top, right, bottom))
```

```python
target_size = (600, 600)
```

```python
import os
```

```python
# assign directory
directory = '/Volumes/PHILIPS UFD/dataset/neg'
 
# iterate over files in that directory
for filename in log_progress(os.listdir(directory), every = 1):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        
        output_path = os.path.join('/Volumes/PHILIPS UFD/dataset_cropped/neg', filename)
        try:
            # Open the input image
            with Image.open(f) as img:
                # Crop image to keep the center
                cropped_img = crop_center(img, *target_size)
                # Save the cropped image as PNG
                cropped_img.save(output_path, "PNG")
                # print(f"Image cropped, saved as {output_path}")
        except IOError:
            print(f"Unable to crop image at {f}")
        except SyntaxError:
            print(f"SyntaxError at {f}")
```
